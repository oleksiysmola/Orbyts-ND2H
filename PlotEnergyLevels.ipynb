```python
import matplotlib.pyplot as plt
import pandas as pd 
import scienceplots
plt.style.use('science')
plt.rcParams['axes.linewidth'] = 2
%matplotlib inline
```

```python
marvelColumns = ["nu1", "nu2", "nu3a", "nu3b", "nu4a", "nu4b", "J", "Ka", "Kc", "inv", "E", "Uncertainty", "transitions"]

marvelEnergies = pd.read_csv("MarvelEnergyLevelsNH2D.txt", delim_whitespace=True)
marvelEnergies["band"] = marvelEnergies["nu1"].astype(str) + marvelEnergies["nu2"].astype(str) + marvelEnergies["nu3a"].astype(str) + marvelEnergies["nu3b"].astype(str) +  marvelEnergies["nu4a"].astype(str) +   marvelEnergies["nu4b"].astype(str)

groundState = marvelEnergies[marvelEnergies["band"] == "000000"]
v1State = marvelEnergies[marvelEnergies["band"] == "100000"]
v2State = marvelEnergies[marvelEnergies["band"] == "010000"]
v3aState = marvelEnergies[marvelEnergies["band"] == "001000"]
v3bState = marvelEnergies[marvelEnergies["band"] == "000100"]
v4aState = marvelEnergies[marvelEnergies["band"] == "000010"]
v4bState = marvelEnergies[marvelEnergies["band"] == "000001"]
```

```python
fontsize=20
markersize=5
plt.figure(figsize=(8, 7))
# print(marvelEnergies.head(20).to_string(index=False))
plt.plot(groundState["J"], groundState["E"], "b.", label="g.s", markersize=markersize)
plt.plot(v1State["J"], v1State["E"], "r.", label=r"$\nu_1$", markersize=markersize)
plt.plot(v2State["J"], v2State["E"], "m.", label=r"$\nu_2$", markersize=markersize)
plt.plot(v3aState["J"], v3aState["E"], "y.", label=r"$\nu_{3_a}$", markersize=markersize)
plt.plot(v3bState["J"], v3bState["E"], "g.", label=r"$\nu_{3_b}$", markersize=markersize)
plt.scatter(v4aState["J"], v4aState["E"], color="darkblue", label=r"$\nu_{4_a}$", s=markersize)
plt.scatter(v4bState["J"], v4bState["E"], color="firebrick", label=r"$\nu_{4_b}$", s=markersize)
plt.xticks(fontsize=fontsize, weight="bold")
plt.yticks(fontsize=fontsize, weight="bold")
plt.ylabel(r"$\mathbf{E/hc}$, $\mathbf{cm}^{-1}$", fontsize=fontsize, fontweight='bold')
plt.xlabel(r"$\mathbf{J}$", fontsize=fontsize, fontweight='bold')
plt.legend(frameon=True, fontsize=fontsize, prop={"size": 15})
plt.savefig("NH2D-EnergiesJ.pdf")
# plt.legend(fontsize=fontsize, prop={"size": 30}, frameon=True)
```
